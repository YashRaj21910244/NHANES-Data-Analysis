## Importing Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
%matplotlib inline
import matplotlib.pyplot as plt
print("Libraries Imported")

pd.set_option('display.max_columns', 100)

path = "nhanes_2015_2016.csv"
```

## Filtering out the Data for Analysis

Now that we have imported the dataset required for our analysis from the .csv file, we will filter out the parameters that we need for our analysis, and we'll also determine the characteristics of these parameters like mean, median, standard deviation, etc.

```python
# Next, look at the 'head' of our DataFrame 'df'. 

head = df.head()
print(head)
```

```python
# Selecting the "BPXSY2" or second systolic blood pressure measurement column as our dataset for analysis
bp = df['BPXSY2']
```

```python
# Mean of the data
bp_mean = np.mean(bp)
print(bp_mean)
bp = bp.dropna()
```

```python
# Median of the data
bp_median = np.median(bp)
print("median =",bp_median)

# Maximum value.ie. Mode of the data
bp_max = np.max(bp)
print("mode =",bp_max)

# Minimum value of the data
bp_min = np.min(bp)
print("min. value =",bp_min)

# Standard Deviation of the data
bp_std = np.std(bp)
print("standard deviation =",bp_std)

# Variance of the data
bp_var = np.var(bp)
print("variance =",bp_var)

# Interquartile Range (IQR) of the data
bp_iqr = stats.iqr(bp) 
print("IQR =",bp_iqr)
```

```python
# We can also get a numerical summary of the selected dataset using the following command
bp_descriptive_stats = bp.describe()
bp_descriptive_stats
```

```python
# Now, making a histogram of out selected dataset
sns.distplot(bp).set(title='BPXSY2')
```

## Data Analysis on Blood Pressure

Now we will carry out further analysis on the data we have filtered out and reach some conclusion using various graphs available in seaborn as well as matplotlib library in Python.

```python
# Now, we will make a boxplot of second systolic and diastolic blood pressure measurements to make a comparison.
bp = sns.boxplot(data=df.loc[:, ["BPXSY2", "BPXDI2"]])
_ = bp.set_ylabel("Blood pressure in mm/Hg")
```

Now we will see the correlation between the systolic and diastolic blood pressure (essentially the maximum and minimum blood pressure between two consecutive heart beats). The correlation coefficient comes out to be 0.32, which signifies a weak correlation between the two values. This means that patients having high systolic blood pressure may not necessarily have high diastolic blood pressure, and vice versa.

Following is a Density Gradient Graph. In the plots below, darker colors indicate where a greater number of points fall. The two plot margins show the densities for the systolic and diastolic blood pressure separately, while the plot in the center shows their density jointly.

```python
sns.jointplot(x="BPXSY1", y="BPXDI1", kind='kde', data=df).annotate(stats.pearsonr)
```

Next we look at two repeated measures of systolic blood pressure, taken a few minutes apart on the same person. These values are very highly correlated, with a correlation coefficient of around 0.96.

```python
jp = sns.jointplot(x="BPXSY1", y="BPXSY2", kind='kde', data=df).annotate(stats.pearsonr)
```

Now in order to find out some more correlations between various parameters in the data, we will carry out further anaylsis.

We will now check how the systolic blood pressure of a person is expected to change with age, and also compare the change in men and women.

```python
#Relation of Systolic Blood Pressure with respect to age of the patient
sns.lineplot(x="RIDAGEYR", y="BPXSY1", data=df)
```

The correlation coefficient comes out to be 0.47, which implies that there is a weak positive correlation between the age of a patient and the Systolic Blood Pressure. Hence, we can conclude that on an average, a person is expected to have a higher blood pressure as his age increases, or in other words, 

<b>Older people tend to have higher Blood Pressure as compared to younger people.</b>

-------------------------------------------------------------------------------------------------------------------------------

```python
#Relation of Systolic Blood Pressure with respect to gender and age of the patient
df["RIAGENDRx"] = df.RIAGENDR.replace({1: "Male", 2: "Female"})
sns.lineplot(x="RIDAGEYR", y="BPXSY1", hue="RIAGENDRx", data=df)
```

We see from the figure above that within each gender, within an age band, the relationship between gender and systolic blood pressure is somewhat complex -- <b>in younger people, men have substantially higher blood pressures than women of the same age. However for people older than 50, this relationship becomes much weaker, and among people older than 70 it appears to reverse.</b> It is also notable that the variation of these distributions, reflected in the height of each box in the boxplot, increases with age.
